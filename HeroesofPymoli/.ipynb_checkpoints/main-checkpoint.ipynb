```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(file_to_load)
```

## Player Count


* Display the total number of players


```python
playercount = data["SN"].nunique()
print(f"There are {playercount} players.")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = data["Item ID"].nunique()
averageprice = data["Price"].mean()
numberofpurchases = data["Purchase ID"].count()
revenue = data["Price"].sum()
items_dict = {'Unique Items': [uniqueitems],
              'Average Purchase Price': [averageprice],
              'Total Number of Purchases': [numberofpurchases],
              'Total Revenue': [revenue],
}
items_df = pd.DataFrame(items_dict)
items_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genderdata = data.groupby(data["Gender"])
gendercount = genderdata["SN"].nunique()
genderperc = gendercount/playercount*100
gender_dict = {'Total Count': gendercount,
               'Percentage': genderperc,
}
gender_df = pd.DataFrame(gender_dict)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchcount = genderdata["Purchase ID"].count()
avgprice = genderdata["Price"].mean()
totalprice = genderdata["Price"].sum()
perperson = totalprice/gendercount
genderanaly_dict = {'Purchase Count': purchcount,
                    'Average Purchase Price': avgprice,
                    'Total Purchase Value': totalprice,
                    'Avg. Purchase Total/Person': perperson
}
genderanaly_df = pd.DataFrame(genderanaly_dict)
genderanaly_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10,15,20,25,30,35,40,100]
binnames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
data["Age Group"] = pd.cut(data["Age"],bins,labels=binnames,right=0)
agedata = data.groupby(data["Age Group"])
agecount = agedata["SN"].nunique()
ageperc = agecount/playercount*100
age_dict = {'Total Count': agecount,
            'Percentage': ageperc,
}
age_df = pd.DataFrame(age_dict)
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchcount = agedata["Purchase ID"].count()
avgprice = agedata["Price"].mean()
totalprice = agedata["Price"].sum()
perperson = totalprice/agecount
ageanaly_dict = {'Purchase Count': purchcount,
                    'Average Purchase Price': avgprice,
                    'Total Purchase Value': totalprice,
                    'Avg. Purchase Total/Person': perperson
}
ageanaly_df = pd.DataFrame(ageanaly_dict)
ageanaly_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenddata = data.groupby("SN")
purchcount = spenddata["Purchase ID"].count()
avgprice = spenddata["Price"].mean()
totalprice = spenddata["Price"].sum()
spendanaly_dict = {'Purchase Count': purchcount,
                    'Average Purchase Price': avgprice,
                    'Total Purchase Value': totalprice,
}
spendanaly_df = pd.DataFrame(spendanaly_dict).sort_values("Total Purchase Value", ascending=0)
spendanaly_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemdata = data[["Item ID", "Item Name", "Price"]]
popdata = itemdata.groupby(["Item ID", "Item Name"])
purchcount = popdata["Item ID"].count()
avgprice = popdata["Price"].mean()
totalprice = popdata["Price"].sum()
popanaly_dict = {'Purchase Count': purchcount,
                    'Item Price': avgprice,
                    'Total Purchase Value': totalprice,
}
popanaly_df = pd.DataFrame(popanaly_dict).sort_values("Purchase Count", ascending=0).head()
popanaly_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitanaly_df = pd.DataFrame(popanaly_dict).sort_values("Total Purchase Value", ascending=0).head()
profitanaly_df.style.format({"Item Price":"${:,.2f}","Total Purchase Value":"${:,.2f}"})
```
